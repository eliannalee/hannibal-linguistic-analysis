## 9. The text’s frequent bigrams and trigrams. Write about the patterns you see among these lists.
***(note: the indexing starts at different/weird places because that is where the syntactical bigrams/trigrams stopped for the most part)***
In both the frequent bigrams and trigrams, many of the word combinations are preceding dialogue or addressing someone, which is pretty standard for novels, however, this could also suggest that much of the plot in Dorian Gray is developed through its dialogue. A common bigram that I think describes most of the novel's subject is ('beautiful', 'things') because the novel focuses primarily on aesthetics and the consumption of things for the simple sake of pleasure. Another interesting bigram is ('you', 'dorian') which shows that a significant amount of dialogue is speaking directly at Dorian Gray; additionally this could show how Lord Henry often gives Dorian advice, which would warrant the use of "you, dorian". A frequent trigram that caught my eye was ('good', 'enough', 'for'), which is not a very common trigram in our daily lexicon; this demonstrates Dorian's constant struggle throughout the novel. Because of his hedonistic nature, and because Dorian gets most everything he desires, the interests that he pursues are never enough to satiate his taste for pleasure. If I recall correctly, there is a whole chapter of the novel that describes Dorian's countless niche interests and endeavors throughout his years and how he was never fully satisfied with what they offered him. A group of trigrams that interested me was ('the', 'world', 'is') and ('the', 'nineteenth', 'century') because Lord Henry often makes commentary about the current state of affairs in the nineteenth century. These two trigrams show the emphasis that the novel places on social critique and commentary.

```python
frequent_bigrams = nltk.FreqDist(nltk.bigrams(dorianText))
list(frequent_bigrams)[57:]
```

```python
frequent_trigrams = nltk.FreqDist(nltk.trigrams(dorianText))
list(frequent_trigrams)[60:]
```

## 8. The text’s collocations. Write about what you notice about the text’s frequently-occurring words.
Many of the novel's collocations are the names of the characters, as their formal names (especially for characters like Lady Narborough and Lady Agatha) are used quite often since Lord Henry and Dorian Gray occupy rather aristocratic spaces. "said Lord" and "said Dorian" are two collocations which show that the majority of the novel's dialogue is said by Dorian and Lord Henry, as Dorian does spend a lot of his time with Lord Henry, learning the ways of socialite London. Additionally, "adrian singleton" is an interesting common collocation, since the actual character is not physically introduced until the end of the novel. Basil Hallward mentions Adrian Singleton in a list of names of people who have had their lives ruined by Dorian's influence; this collocation demonstrates the importance of Adrian Singleton's presence as an example of Dorian Gray's hedonistic and overall harmful influence. As I wrote for the previous prompt about the trigrams, the commonly occurring phrase of "nineteenth century" demonstrates the emphasis that this novel places on the time period and the atmosphere that the late nineteeth century brings with it.
